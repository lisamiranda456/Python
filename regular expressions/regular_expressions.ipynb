# Regular Expressions

### List of characters with their meanings
### ^ - matches beginning of the line
### $ -matches end of the line


### . - matches any single character except newline. Using option m,then newline can also be matched.
### [...] - Matches any single character in bracket.
### [^...] - Matches any single character NOT in brackets.
### exp* - Matches 0 or more occurrences of preceeding expression.
### exp+ - Matches 1 or more occurrences of preceeding expression.
### exp? - Matches 0 or 1 occurrence of preceeding expression.
### exp{ n} -Matches exactly n number of occurrences of preceeding expression.
### exp{ n, } -Matches n  or more number of occurrences of preceeding expression.



### exp{n,m} - Matches at least n and at most m occurrences of preceeding expression.
### a| b - Matches either a or b.
### (exp) - groups regular expressions and remembers matched text.
### \d -Matches digits. Equivalent to [0-9].
### \D -Matches non digits.
### \w - Matches word characters.
### \W - Matches non-word characters.
### \s - Matches whitespace.Equivalent to [\t\n\r\f].
### \S - Matches non- whitespace .
### \A -Matches beginning of String.
### \Z - Matches end of string. If a newline exists, it matches just before newline.
### \z - Matches end of string.



### [Pp]ython - Match "Python" or "python".
### rub[ye] - Match ruby or rube.
### [aeiou] - Matches any one lower case vowel.
### [0-9] - Match any digit . same as [0123456789].
### [a-z] -Match any lower case ACSII letter.
### [A-Z] -Match any upper case ASCII letter.
### [a-zA-Z0-9] - Match any of uppercase , lower case alphabets and digits.
### [^aeiou] -Match anything other than lower case vowel.
### [^0-9] -Match anything other than a digit.


```python
#simple example of regex
##display the lines which are containing how
import re
fhand=open("myfile.txt",'r')
for line in fhand:
    line=line.rstrip()
    if re.search("how",line):
        print(line)
fhand.close()
```

```python
#simple example of regex
##display the lines which are containing how
import re
fhand=open("myfile.txt",'r')
for line in fhand:
    line=line.rstrip()
    if re.search("^how",line):
        print(line)
fhand.close()
```

```python
#line starting with I followed by two characters then follwed by m, length should be atleast four characters 
import re
fhand=open("myfile.txt",'r')
for line in fhand:
    line=line.rstrip()
    if re.search("^I..m",line):
        print(line)
fhand.close()
```

```python
#lines starting with "h" and followed one or more characters and containing u.
import re
fhand=open("myfile.txt",'r')
for line in fhand:
    line=line.rstrip()
    if re.search("^h.+u",line):
        print(line)
fhand.close()
```

```python
#lines starting with h and followed by 0 or more characters and containing u.
import re
fhand=open("myfile.txt",'r')
for line in fhand:
    line=line.rstrip()
    if re.search("^h.*u",line):
        print(line)
fhand.close()
```

```python
#Write a program to extract lines starting with the word "From" or "from" 
#and having mail id with .edu
#input file is log.txt
import re
count1=0
fhand=open("log.txt",'r')
for line in fhand:
    count1=count1+1
    line=line.rstrip()
    if re.search("^[Ff]rom.*edu$",line):
        
        print("line number",count1,line)
fhand.close()
```

```python
#print only domain names eg umich.edu
import re
s="A message from csev@umich.edu to cwen@iupui.edu about meeting @2PM"

lst=re.findall('@[a-z]\S+',s)
for i in lst:
    print(i[1:])
```
